# **1. Importar e Preparar os Dados**

 Importar os dados para um DataFrame no Colab e preparar para análise.


```python
# Importando a Biblioteca do Pandas
import pandas as pd
```

```python
# Criando o DataFrame com os dados fornecidos
dados = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df_relatorio = pd.DataFrame(dados)

# Verificando a tabela
display(df_relatorio)
```

```python
# Conferindo as colunas da tabela
df_relatorio.columns
```

# **2. Análise Exploratória dos Dados**
 Calcular as métricas solicitadas:


*   total e média diária de horas trabalhadas
*   bugs corrigidos
*   tarefas concluídas

Para fazer os calculos iremos utilizar o método *.sum()* para realizar a soma dos elementos do dataframe. E tambem utilizamos o método *.mean()*, para calcular a média de valores numéricos, e depois usamos *round()* para arredondar o resultado para o número inteiro mais próximo. Isso simplifica os números para que sejam mais fáceis de entender e usar posteriormente.

### Variáveis
Definindo as variáveis das colunas para armazenar os resultados:

*   total e média diária de horas  
  *   total_horas = df['Horas Trabalhadas']
  *   media_horas = df['Horas Trabalhadas']
<br>

*   bugs corrigidos
  *   total_bugs = df['Bugs Corrigidos']
  *   media_bugs = df['Bugs Corrigidos']
<br>
  
*   tarefas concluídas
  *   total_tarefas = df['Tarefas Concluídas']
  *   media_tarefas = df['Tarefas Concluídas']



```python
# Total de Horas Trabalhadas
total_horas = df_relatorio['Horas Trabalhadas'].sum()
media_horas = df_relatorio['Horas Trabalhadas'].mean().round()

print("Total de Horas Trabalhadas:", total_horas)
print("Média Diária de Horas Trabalhadas:", media_horas)
```

```python
# Total de Bugs Corrigidos
total_bugs = df_relatorio['Bugs Corrigidos'].sum()
media_bugs = df_relatorio['Bugs Corrigidos'].mean().round()

print("Total de Bugs Corrigidos:", total_bugs)
print("Média Diária de Bugs Corrigidos:", media_bugs)
```

```python
# Total de Tarefas Concluídas
total_tarefas = df_relatorio['Tarefas Concluídas'].sum()
media_tarefas = df_relatorio['Tarefas Concluídas'].mean().round()

print("Total de Tarefas Concluídas:", total_tarefas)
print("Média Diária de Tarefas Concluídas:", media_tarefas)
```

### Calcular a Produtividade Diária (Tarefas Concluídas por Hora)

Para calcular a produtividade diária (tarefas concluídas por hora), criamos uma nova coluna chamada “Produtividade” no DataFrame “df”, onde cada valor é o resultado da divisão dos valores da coluna "Tarefas Concluídas" pelos valores da coluna "Horas Trabalhadas".
<br>

Vamos calcular a média dos valores da nova coluna Produtividade para obter a média diária de “Produtividade” .




```python
# Produtividade Diária (Tarefas Concluídas por Hora)
df_relatorio['Produtividade'] = df_relatorio['Tarefas Concluídas'] / df_relatorio['Horas Trabalhadas']
media_produtividade = df_relatorio['Produtividade'].mean().round()

print("Média Diária de Produtividade:", media_produtividade)
```

Um novo DataFrame produtividade_diaria é criado para exibir os dias da semana, as tarefas concluídas e a produtividade diária. <br>
Os valores são arredondados para melhorar a legibilidade.

```python
produtividade_diaria = df_relatorio [['Dia','Tarefas Concluídas','Produtividade']].round()
display(produtividade_diaria)
```

# 3. Visualização dos Dados

Criando os gráficos para representar os dados e os insights.

```python
# Utilizamos a biblioteca matplotlib.pyplot para criar diferentes tipos de gráficos.
import matplotlib.pyplot as plt
```

```python
# Utilizamos a biblioteca seaborn que oferece uma interface mais simples e
# elegante para criar gráficos estatísticos, baseada no matplotlib.
import seaborn as sns
```

```python
# Configura o estilo dos gráficos usando seaborn para
# criar uma grade de fundo branca, que ajuda a destacar os dados.
sns.set(style="whitegrid")
```

Cada gráfico é configurado para representar diferentes métricas ao longo dos dias da semana, proporcionando uma visualização clara e compreensível dos dados analisados.

```python
# Gráfico de Horas Trabalhadas por Dia
plt.figure(figsize=(10, 5))
sns.barplot(x='Dia', y='Horas Trabalhadas', data=df_relatorio, palette='viridis')
plt.xlabel('Dia da Semana')
plt.ylabel('Horas Trabalhadas')
plt.title('Horas Trabalhadas por Dia')
plt.xticks(rotation=45)
plt.show()
```

*   *plt.figure(figsize=(10, 5)):* Configura o tamanho do gráfico.
* *sns.barplot(...):* Cria um gráfico de barras. x='Dia' define o eixo x (dias da semana) e y='Horas Trabalhadas' define o eixo y (horas trabalhadas).
*   *palette='viridis':* Define a paleta de cores para o gráfico.
*   *plt.xlabel/plt.ylabel/plt.title:* Adiciona rótulos aos eixos e um título ao gráfico.
*   *plt.xticks(rotation=45):* Rotaciona os rótulos do eixo x em 45 graus para melhor legibilidade.
*   *plt.show():* Exibe o gráfico.






```python
# Gráfico de Barras para Bugs Corrigidos
plt.figure(figsize=(10, 5))
sns.barplot(x='Dia', y='Bugs Corrigidos', data=df_relatorio, palette='viridis')
plt.xlabel('Dia da Semana')
plt.ylabel('Bugs Corrigidos')
plt.title('Bugs Corrigidos por Dia')
plt.xticks(rotation=45)
plt.show()
```

*   Semelhante ao gráfico anterior, mas o eixo y agora representa o número de bugs corrigidos.


```python
# Gráfico de Barras para Tarefas Concluídas
plt.figure(figsize=(10, 5))
sns.barplot(x='Dia', y='Tarefas Concluídas', data=df_relatorio, palette='viridis')
plt.xlabel('Dia da Semana')
plt.ylabel('Tarefas Concluídas')
plt.title('Tarefas Concluídas por Dia')
plt.xticks(rotation=45)
plt.show()
```

*   Semelhante aos gráficos anteriores, mas o eixo y agora representa o número de tarefas concluídas.



```python
# Gráfico de Linha para Produtividade Diária
plt.figure(figsize=(10, 5))
sns.lineplot(x='Dia', y='Produtividade', data=df_relatorio, marker='o', color='blue')
plt.xlabel('Dia da Semana')
plt.ylabel('Produtividade (Tarefas por Hora)')
plt.title('Produtividade por Dia')
plt.xticks(rotation=45)
plt.show()
```


*   sns.lineplot(...): Cria um gráfico de linha. x='Dia' define o eixo x (dias da semana) e y='Produtividade' define o eixo y (produtividade diária).
*  marker='o': Adiciona marcadores circulares nos pontos de dados.
*  olor='blue': Define a cor da linha.
